```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generates a numpy array of 10 rows and 1 column
x = np.random.random((10,1))
x
```

```python
# If we plot only an array the x axis would be the indexes and the y axis would be the values
plt.plot(x,'*-')
plt.show();
```

```python
x = np.linspace(0,10,100)
y = np.power(x,0.5)
plt.plot(x,y,'*')
plt.show();
```

```python
 import seaborn as sns
 sns.set()
 sns.lineplot(x,y)
 plt.show();
```

```python
import pandas as pd
mobile_data = pd.read_csv("https://raw.githubusercontent.com/likhitha9reddy/PadhAI/master/mobile_cleaned.csv")
mobile_data.head()
```

```python
ax = sns.scatterplot(x="stand_by_time",y="battery_capacity", data = mobile_data)
```

```python
ax = sns.scatterplot(x="stand_by_time",y="battery_capacity", hue="thickness", data=mobile_data)
```

**Histogram**

```python
ax = sns.distplot(mobile_data["stand_by_time"])
```

```python
ax = sns.distplot(mobile_data["stand_by_time"],kde=False,rug=True)
```

Increase the number of bins to understand better

```python
ax = sns.distplot(mobile_data["stand_by_time"],kde=False,rug=True,bins=20)
```

```python
ax = sns.boxplot(x="is_liked",y="battery_capacity",data=mobile_data)
```

```python
ax = sns.boxplot(x="expandable_memory",y="price",data=mobile_data)
```

```python
uniform_data = np.random.rand(10,12)
print(uniform_data)
```

```python
# We have 10 rows and 12 columns
ax = sns.heatmap(uniform_data)
```

```python
ax = sns.heatmap(uniform_data,cmap='YlGnBu')
```

```python
ax = sns.pairplot(mobile_thin, diag_kind='hist', hue='is_liked')
```
