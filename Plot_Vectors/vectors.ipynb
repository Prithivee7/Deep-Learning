```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.quiver(0,0,3,4)
```

```python
plt.quiver(0,0,3,4, scale_units='xy', angles='xy', scale=1)
plt.show();
```

```python
# Vector from (0,0) to (3,4)
plt.quiver(0,0,3,4, scale_units='xy' , angles='xy', scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show();
```

```python
def plot_vectors(vectors):
  colors = ['r', 'b', 'g', 'y']
  i = 0
  for vector in vectors:
    plt.quiver(vector[0], vector[1], vector[2], vector[3], scale_units='xy', angles='xy', scale= 1, color=colors[i % len(colors)])
    i += 1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show();
```

```python
vectors = [np.asarray([0,0,3,4]),np.asarray([0,0,-4,4]),np.asarray([0,0,-8,-4]),np.asarray([0,0,3,-4])]
plot_vectors(vectors)
```

### Adding the vectors

```python
plot_vectors([vectors[0], vectors[2], vectors[0] + vectors[2]])
```

```python
plot_vectors([vectors[0], vectors[3], vectors[0] + vectors[3]])
```

# Subtracting the vectors. 

For this we take the negative of a vector

```python
plot_vectors([vectors[0], -vectors[3], vectors[0] + -vectors[3]])
```

### Dot product of vectors

```python
a = np.asarray([3,4])
b = np.asarray([4,-1])
```

$\vec{a}\cdot\vec{b} = |\vec{a}| |\vec{b}| \cos(\theta) = a_x b_x + a_y b_y$

```python
a_dot_b = np.dot(a,b)
a_dot_b
```

### Projection of a on b

$ a_b=|\vec{a}| \cos(\theta) = |\vec{a}|\frac{\vec{a}\cdot\vec{b}} {|\vec{a}| |\vec{b}|}  = \frac{\vec{a}\cdot\vec{b}} {|\vec{b}|}$

We get this by substituting $cos(\theta)$ from the above equation

```python
# Projection of a on b
# from a_b we will get the magnitude

a_b = np.dot(a,b) / np.linalg.norm(b)
a_b
```
